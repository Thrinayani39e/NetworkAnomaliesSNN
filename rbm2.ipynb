```python
import numpy as np
import pandas as pd
from sklearn.neural_network import BernoulliRBM
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load training and test data from CSV files
train_data = pd.read_csv('train_set_after_pca.csv')
test_data = pd.read_csv('test_set_after_pca.csv')
train_data.dropna(inplace=True)
# Separate features and labels
X_train = train_data.iloc[:, :-1].values
y_train = train_data.iloc[:, -1].values
X_test = test_data.iloc[:, :-1].values
y_test = test_data.iloc[:, -1].values

# Preprocess the data (Standardize features)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

```python
from sklearn.neural_network import BernoulliRBM
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize and train the RBM with modified hyperparameters
rbm7 = BernoulliRBM(n_components=256, learning_rate=0.01, n_iter=30, verbose=0)
rbm7.fit(X_train)

# Transform the data using the RBM
X_train_transformed = rbm7.transform(X_train)
X_test_transformed = rbm7.transform(X_test)

# Train a Logistic Regression classifier with modified hyperparameters
classifier = LogisticRegression(max_iter=1000, C=0.1)  # Modify C and other parameters as needed
classifier.fit(X_train_transformed, y_train)

# Predict on the test data
y_pred = classifier.predict(X_test_transformed)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
import joblib
joblib.dump(rbm7, '/content/drive/My Drive/Final year project/models/rbm7.keras')
```
